```python
"""Softmax."""

"""Old code: Tracer()()"""

scores = [3.0, 1.0, 0.2]

import numpy as np
import pdb;

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    exp_all = np.exp(x)
    sum_all = np.sum(exp_all,axis=0)
    return exp_all/sum_all


print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
#pdb.set_trace()
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
pdb.set_trace()

plt.plot(x, softmax(scores).T, linewidth=3)
plt.show()
```
